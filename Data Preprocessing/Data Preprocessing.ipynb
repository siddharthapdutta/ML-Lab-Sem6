```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Data1.csv')
print(dataset.head())
```

Independent Variables: R&D Spend, Administration, Marketing Spend, State

Dependent Variable: Profit

```python
# Extracting all rows and columns from the given dataset
dataset_1 = dataset.iloc[:,:]
print(dataset_1)
```

```python
# Extracting from Row Number 0 to Row Number 10
# Extracting From Column Number 0 to Column Number 2
dataset_2 = dataset.iloc[:11,:3]
print(dataset_2)
```

```python
# Extracting from Row Number 12 to Last Row
# Extracting from Column Number 1 to Last Column
dataset_3 = dataset.iloc[12:,1:]
print(dataset_3)
```

```python
# Dividing dataset into independent variables and dependent variables
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(x)
print(y)
```

```python
# Data Pre-Processing: Missing Values
from sklearn.preprocessing import Imputer
impute = Imputer(strategy="mean", axis=0) # Default Strategy: Mean
x[:, 0:3] = impute.fit_transform(x[:,0:3])
print(x)
```

Mean Value for Column 0: 75769.86521739131

Mean Value for Column 1: 121228.07693877546

Mean Value for Column 2: 212626.22434782612

```python
impute = Imputer(strategy="median", axis=0) # Strategy: Median
x[:, 0:3] = impute.fit_transform(x[:,0:3])
print(x)
```

Median Value for Column 0: 74661.715

Median Value for Column 1: 122616.84

Median Value for Column 2: 212716.24


```python
impute = Imputer(strategy="most_frequent", axis=0) # Strategy: Most Frequent
x[:, 0:3] = impute.fit_transform(x[:,0:3])
print(x)
```

Most Frequent Value for Column 0: 542.05

Most Frequent Value for Column 1: 51283.14

Most Frequent Value for Column 2: 0.0

```python
# Labelling Categorical Data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 3] = le.fit_transform(x[:, 3])
print(le.classes_)
```

Classes of Label Encoder: ['California' 'Florida' 'New York']

Mapping: {‘California’: 0, ‘Florida’: 1, ‘New York’:2}

```python
# Performing One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
countries = enc.fit_transform(x[:, 3].reshape(-1, 1)).toarray()
x = x[:, :-1]
x = np.concatenate((x, countries), axis=1)
print(x)
```


Mapping: {2:[0,0,1], 0:[1,0,0],1:[0,1,0]}


```python
# Reducing Dimensions
# (50, 6) -> (50, 5)
x = x[:,:-1]
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
print(x)
```
